```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Código para gerar impulso

array = np.arange(-100, 101)
l = np.size(array)
impulse = np.zeros(l)
indice = np.where(array==5)
impulse[indice] = 1

plt.stem(array, impulse, linefmt = 'k--')
plt.title("Impulso")
plt.xlabel("Amostras")
plt.ylabel("Amplitude")
plt.grid()
```

```python
# Código para gerar degrau unitário

array = np.arange(-100, 101)
l = np.size(array)
step = np.zeros(l)
indice = np.where(array>=25)
step[indice] = 1

plt.stem(array, step, linefmt = 'k--')
plt.title("Degrau")
plt.grid()
```

```python
#Código para gerar Sequência Sinusoidal

Fs = 8000 # Frq d/ amostragem
f0 = 1000 # Frequência (Determina distância da sin)
t = 100 # tempo de exec ms
A = 1000 # valor Amplitude

t = np.arange(t)
y = np.cos(2 * np.pi * f0 * (t / Fs))
y = A * y

plt.stem(t, y, linefmt = 'k--')
plt.title("Sequência Sinusoidal")
plt.grid()
```

```python
#Código para gerar Sequência Exponencial

A = 10000 #Amplitude
n = np.arange(100)
a1 = 0.5

saida = np.zeros(np.size(n))

for i in range(len(n)):
  saida[i] = (A*(a1**i)) # 2^i

plt.stem(n, saida, linefmt = 'k--')
plt.title("Sequência Exponencial")
plt.grid()
```
