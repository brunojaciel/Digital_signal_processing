```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.00005
t = np.arange(-0.005, 0.005, dt)
f = 2000

xa = np.cos(2 * np.pi * f * t)
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(t*1000, xa)
ax.grid()
ax.set_title(f'Sinal de Freq {f} Hz')
ax.set_xlabel('Tempo [ms]')
ax.set_ylabel('xa(t)')
```

```python
Fs = 8000
Ts = 1/Fs
n = np.arange(-20, 20, 1)
xd = np.cos(2*np.pi*n*(f/Fs))

fig, ax = plt.subplots(figsize=(10, 4))
ax.stem(n*Ts*1000, xd, linefmt = 'r--')
ax.grid()
ax.set_title('Sinal amostrado')
ax.set_xlabel('Amostras')
ax.set_ylabel('x[n]')
```
