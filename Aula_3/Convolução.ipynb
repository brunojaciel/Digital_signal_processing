```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.ones(6)
h = np.zeros(4)
h[0] = 1
h[1] = 0.5
h[2] = 0.25
h[3] = 0.125
```

```python
array = np.arange(len(x))

plt.stem(array, x)
```

```python
array = np.arange(len(h))
plt.stem(array, h)
```

```python
convolution = np.convolve(x, h)
```

```python
array = np.arange(np.size(convolution))
plt.stem(array, convolution, )
```
