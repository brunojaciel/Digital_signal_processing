```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Código para gerar degrau unitário

array = np.arange(-30, 31)
l = np.size(array)
step = np.zeros(l)
print(array)
indice = np.where((array>=0) & (array<=2))
print(indice)
step[indice] = 1
plt.stem(array, step)
plt.title("Degrau")
plt.grid()
```

```python
M = 16 # Tamanho

x = np.zeros((M, 1))
coef = np.zeros((M, 1))

for j in range(0, M):
  coef[j] = 1/M
```

```python
sav_y = np.zeros((l, 1))

for j in range(0, l):
  x[0] = step[j]

  y = 0
  for n in range(0, M):
    y = y + coef[n] * x[n]

  sav_y[j] = y
  for n in range(M-1, 0, -1):
    x[n] = x[n-1]


plt.stem(array, sav_y)
plt.grid()
```
