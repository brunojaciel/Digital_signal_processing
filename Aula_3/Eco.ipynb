```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Código para gerar degrau unitário

array = np.arange(-100, 101)
l = np.size(array)
step = np.zeros(l)
#indice = np.where((array>=0) & (array<=2))
indice = np.where(array==0)
step[indice] = 1
plt.stem(array, step)
plt.title("Degrau")
plt.grid()
```

```python
Fs = 8000 # Frq de amostragem
t1 = 20*10**-3 # tempo 1 para o n1
#t2 = 15*10**-3 # tempo 2 para o n2
a0 = 0.9 # amostra 0
a1 = 0.7 # amostra 1
#a2 = 0.2 # amostra 2
n1 = int(t1*Fs) #
#n2 = int(t2*Fs) #
print(n1)

y = np.zeros(np.size(step))

for i in range(np.size(step)):
    #Verificar o N1
    #y[i] = (a0 * step[i] + a1 * y[i - n1] )#+ a2 * y[i - n2])
    y[i] = (a0 * step[i] + a1 * y[i - 10] )
plt.stem(array, y)
plt.title("Eco")
plt.grid()
```
