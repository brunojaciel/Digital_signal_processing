```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 10000
t1 = 0.002
t2 = 0.004
# Exemplo delay
n1 = int(t1*Fs)
n2 = int(t2*Fs)
# Definição do ganho
a0 = 0.5
a1 = 0.3
a2 = 0.2
tama_delay = n2
vetor_delay = np.zeros(int(tama_delay+1))
```

```python
# Código para gerar degrau unitário

array = np.arange(-100, 101)
l = np.size(array)
step = np.zeros(l)
indice = np.where(array==0)
step[indice] = 1
plt.stem(array, step)
plt.title("Degrau")
plt.grid()
```

```python
entrada = step
tama_loop = np.size(step)
vet_saida = np.zeros(tama_loop)

print(n2)

for j in range(0, tama_loop):
  input = entrada[j]
  vetor_delay[0] = input
  y = a0 * vetor_delay[0] + a1 * vetor_delay[n1] + a2 * vetor_delay[n2]
  # Desloca o vetor de delay
  for k in range(0, tama_delay):
    vetor_delay[tama_delay-k] = vetor_delay[tama_delay-k-1]
  vet_saida[j] = y

plt.stem(array, vet_saida)
plt.title("Delay")
plt.grid()
```
