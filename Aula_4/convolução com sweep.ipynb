```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
swerp = np.memmap('/content/drive/MyDrive/Processamento_Digital_De_Sinais/Aula_4/Swerp_40_3k4.pcm', mode='r', dtype=np.int16)
```

```python
array = np.arange(np.size(swerp))
plt.stem(array, swerp)
print(type(swerp))
```

```python
M = 64

coef = np.zeros((M, 1))
x = np.zeros((M, 1))

for j in range(0, M):
  coef[j] = 1/M
```

```python
sav_y = np.zeros((np.size(swerp), 1))

for j in range(0, np.size(swerp)):
  x[0] = swerp[j]

  y = 0
  for n in range(0, M):
    y = y + coef[n] * x[n]

  sav_y[j] = y
  for n in range(M-1, 0, -1):
    x[n] = x[n-1]
```

```python
array = np.arange(np.size(swerp))
plt.stem(array, sav_y)
plt.grid()
```

```python
from scipy.io.wavfile import write

write('sinal_saida.pcm', 8000, sav_y.astype(np.int16))
```
