```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
s = np.memmap('/content/drive/MyDrive/Processamento_Digital_De_Sinais/Aula_Apresentação/Anexo_38986928.pcm', mode='r', dtype=np.int16)

print(type(s))

itera = len(s)

print(f'Size of array:{itera}')
```

```python
array = np.arange(itera)
plt.stem(array, s, linefmt = 'r--')
plt.grid()
plt.title('Sinal de entrada')
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
# Confirmar se pode usar operador ** para plotar
ax.plot(22500, 80000, s, color="C2")
#ax = plt.plot(22500, 80000,s)
ax.grid()
ax.set_title('Sinal de entrada')
```

```python
#Salvar valores intermediários
sav_y = np.zeros(itera)
print(sav_y)
ganho = 0.5

#Executa o processamento

for j in range(1, itera):
  x = s[j]
  y = ganho*x
  sav_y[j] = y

print(x)
print(y)
print(sav_y)
```

```python
plt.stem(array, sav_y, linefmt = 'r--')
plt.grid()
plt.title('Sinal de entrada')
plt.xlabel('Número de amostras')
plt.ylabel('Amplitude da saída')
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(22500, 80000, sav_y, color="C9")
ax.grid()
ax.set_title('Sinal de Saída')
ax.set_xlabel('Número de amostras')
ax.set_ylabel('Amplitude da saída')
```
